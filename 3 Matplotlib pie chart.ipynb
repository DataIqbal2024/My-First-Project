```python
%matplotlib inline
import matplotlib.pyplot as plt
```

# draw pie chart 

```python
exp_vals = [1400,600,300,410,250]
exp_labels = ["Home Rent", "Food", "Phone/Internet Bill", "Car", "Other Utilities"]
plt.pie(exp_vals,labels=exp_labels)
```

# PIE CHART FOR SHADOW

```python
plt.pie(exp_vals, labels=exp_labels, shadow=True)
plt.axis("equal")
plt.show()
plt.savefig("1.jpg")
```

```python
plt.axis("equal")
plt.pie(exp_vals,labels=exp_labels, shadow= True, autopct = '%1.2f%%', radius=1.5)
plt.show()
```

# EXPLODE

```python
plt.axis("equal")
plt.pie(exp_vals,labels=exp_labels, shadow = True, autopct='%1.1f%%', radius=1.5, explode=[0,0,0,0.1,0.9])
plt.show()
```

```python
plt.axis("equal")
plt.pie(exp_vals,labels=exp_labels, shadow = True, autopct='%1.1f%%', radius=1.5, explode=[0,0,0,0.1,0.6],counterclock=True,startangle=45)
plt.show()
```
