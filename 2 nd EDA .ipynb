# Cleaning Data
Cleaning data is crucial step before EDA as it will remove the ambigious

. Remove duplicate rows\
. Handling missing value\
. Convert columns to appropriate datatypes\
. Adding important columns

```python
# 1 Removing duplicate rows if any 

df1.duplicated().sum() # show number of rows of dupliate rows duplicate rows
```

```python
df1[df1.duplicated()].shape
```

```python
# dropping duplicate value

df1.drop_duplicates(inplace = True)
```

```python
df1.shape
```

```python

# 2 -- Handling missing value

# columns having missing value

df1.isnull().sum().sort_values(ascending = False)[:6]
```

```python
df1[['company','agent']] = df1[['company','agent']].fillna(0)
```

```python
df1['children'].unique()
```

```python
df1['children'].fillna(df1['children'].mean(), inplace = True)
```

```python
df1['country'].fillna('others',inplace = True)
```

```python
df1.isna().sum().sort_values(ascending = True)
```

```python
df1[df1['adults']+df1['babies']+df1['children']==0].shape
```

```python
df1.drop(df1[df1['adults']+df1['babies']+df1['children'] ==0].index,inplace = True)
```

# Converting columns to appropriate datatypes

```python
# Converting datatypes of column children 
df1[['children','comany','agent']] = df1[['children','company','agent']].astype('int64')
```

```python
# changing datatype of columns reservation_status_date to data_type

df1['reservation_status_date'] = pd.to_datetime(df1['reservation_status_date'])
```

# Adding important columns

```python
df1['total_stay'] = df1['stays_in_weekend_nights']+df1['stays_in_weekend_nights']

# adding total people num as columns .i . e. total people num = num od adults + children + babies 
df1['total_people'] = df1 ['adults']+df1['children']+df1['babies']
```
