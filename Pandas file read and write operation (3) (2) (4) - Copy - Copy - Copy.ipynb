# Web APIs

```python
import pandas as pd
```

```python
import requests
url = 'https://api.github.com/repos/pandas-dev/pandas/issues'
resp = requests.get(url)
resp
```

```python
data = resp.json()

len(data)
```

```python
data
```

```python
new_data=pd.DataFrame(data)
```

```python
new_data
```

```python
df= pd.DataFrame(data, columns=['number',"user", 'title','labels', 'reactions'])
df.head()                              
```

```python
df["labels"][0]
```

```python
q=[]
for i in df["labels"]:
    if len(i)==0:
        continue
    else:
        for j in i:
            q.append(j)
   
```

```python
q
```

```python
label_new=pd.DataFrame(q)
```

```python
label_new
```
