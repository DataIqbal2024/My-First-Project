# Data Visualisation

```python
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
import pandas as pd
```

```python
plt.style.available
```

```python
x = np.linspace(0,10,100)
x
```

```python
y = np.sin(x)
y
```

```python
plt.style.use("classic")
```

```python

plt.figure(figsize=(23,6))
plt.plot(x,y)
```

```python
plt.figure(figsize=(20,5))

plt.plot(x,y,"--") # sinx

plt.plot(x,np.cos(x))  #cosx
```

```python

plt.subplot()
```

```python
c = plt.figure(figsize=(18,10))
plt.subplot(3,1,1) # 1 row 3 column 1 graph position
plt.plot(x,np.sin(x));

plt.subplot(3,1,2) 
plt.plot(x,np.cos(x));

plt.subplot(3,1,3) 
plt.plot(x,np.tan(x))
```

```python
c = plt.figure(figsize=(10,5))
plt.subplot(1,3,1) # 1 row 3 column 1 graph position
plt.plot(x,np.sin(x));

plt.subplot(1,3,2) 
plt.plot(x,np.cos(x));

plt.subplot(1,3,3) 
plt.plot(x,np.tan(x))
```

```python
data=pd.read_table("http://archive.ics.uci.edu/ml/machine-learning-databases/yeast/yeast.data",sep="\s+",header=None)
data
```

```python
data.columns=["Sequence Name","mcg","gvh","alm","mit","erl","pox","vac","nuc","class"]
```

```python
data
```

```python
data["mcg"].value_counts() # histogram for single series
```

```python
plt.hist(data["mcg"],bins = 10)
plt.savefig("Iqbal.jpg")
```

```python
plt.grid(color="b",linstyle="--")
```

```python
plt.figure(figsize=(12,4))

plt.hist(data.mcg,color="yellow",edgecolor="black")
plt.grid(color="g",linestyle="--")
```
